𝗤𝘂𝗲𝘀𝘁𝗶𝗼𝗻 1:
Write a python program that takes a string as an input from the user. Then 
check which email domain the user is using and print it. For this question, 
assume you have only 4 email domains available and they are gmail, yahoo, 
outlook, and bracu.

[You are NOT allowed to use split() and sting slicing for this task]

================================
Sample Input 1:
tom@gmail.com
Sample Output 1:
The user is using gmail as a mailing domain.
Explanation 1:
Since gmail domain is being used by the user.
==============================================================
Sample Input 2:
kelvin@yahoo.com
Sample Output 2:
The user is using yahoo as a mailing domain.
Explanation 2:
Since yahoo domain is being used by the user.
==============================================================

```python
input = str(input("Please enter a string: "))
var = list(input)
x = ""
for n in range(0,len(input)):
  if var[n] == '@':
    if var[n + 1] == 'g':
      x = x +'The user is using gmail as a mailing domain'
    if var[n + 1] == 'y':
      x = x +'The user is using yahoo as a mailing domain'
    if var[n + 1] == 't':
      x = x +'The user is using tutanota as a mailing domain'
    if var[n + 1] == 'b':
      x = x +'The user is using brac as a mailing domain'
print(x)
```
